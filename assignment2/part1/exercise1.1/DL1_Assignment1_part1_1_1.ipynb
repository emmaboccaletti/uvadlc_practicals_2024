```python
file_paths = ['train_CNN_valid_output.txt', 'train_CNN_replicate_output.txt', 'train_CNN_reflect_output.txt',
 'train_CNN_circular_output.txt', 'train_CNN_sconv_output.txt', 'train_CNN_fconv_output.txt',
 'Net2_train_CNN_valid_output.txt', 'Net2_train_CNN_replicate_output.txt', 'Net2_train_CNN_reflect_output.txt',
 'Net2_train_CNN_circular_output.txt', 'Net2_train_CNN_sconv_output.txt', 'Net2_train_CNN_fconv_output.txt']

new_file_paths = ['outputs/' + file for file in file_paths]

all_results = []

for file_path in new_file_paths:
    with open(file_path, 'r') as file:
        data = file.readlines()

    # Network Type
    network_type = 'Net2' if 'Net2' in file_path else 'Net1'

    # Convolution Type
    convolution_type = next(line for line in data if "Type of convolution" in line).split(':')[1].strip()

    # Extract results for validation and test datasets
    validation_results_line = next(line for line in data if "Results for validation dataset" in line)
    test_results_line = next(line for line in data if "Results for test dataset" in line)

    validation_results = eval("{" + validation_results_line.split('{')[1].split('}')[0] + "}")
    test_results = eval("{" + test_results_line.split('{')[1].split('}')[0] + "}")

    for round_number in validation_results.keys():
        if round_number == 0:
            all_results.append({
                "Network Type": network_type,
                "Convolution Type": convolution_type,
                "Round": round_number + 1,
                "Validation Accuracy": validation_results[round_number],
                "Test Accuracy": test_results[round_number]
            })
        else:
            all_results.append({
                "Network Type": '',
                "Convolution Type": '',
                "Round": round_number + 1,
                "Validation Accuracy": validation_results[round_number],
                "Test Accuracy": test_results[round_number]
            })

final_results_df = pd.DataFrame(all_results)
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)  # Adjust the display width to avoid line wrapping
pd.set_option('display.max_colwidth', None)  # Display full column content

# Display the final table
print(final_results_df)
```
